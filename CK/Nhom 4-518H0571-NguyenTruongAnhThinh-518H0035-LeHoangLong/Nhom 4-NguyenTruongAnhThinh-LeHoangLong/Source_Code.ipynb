```python
import random
import sys
import networkx as nx
import matplotlib.pyplot as plt
```

```python
link ='/content/facebook.txt'
data=open(link, "r")
data_temp=data.readlines()
data_temp
```

```python
print(len(data_temp))
```

```python
data=[]
for row in data_temp:
  temp=row.split()
  if(len(temp) ==2):
    data.append(row)
  
data
```

```python
print(len(data))
```

```python
graph = nx.Graph()
for row in data:
  list_node = row.split()
  # print(list_node)
  # break;
  graph.add_nodes_from(list_node)
  node0=list_node[0]
  node1=list_node[1]
  graph.add_edge(node0, node1)

print("Number of Nodes: ", graph.number_of_nodes())
print("Number of Edges: ", graph.number_of_edges())

```

```python
nx.draw(graph, with_labels=True)
plt.axis([0.1, 0.01, 0.1, 0.01])
plt.show()
```

```python
##Tm cc node lin k ca 1 node
def neighbors_nodes(G, node):
  return [nei for nei in G.neighbors(node)]

list_neighbor_node = neighbors_nodes(graph, "0")
print(list_neighbor_node)
```

```python
##Tm cc nodes lin k nodes ca 1 node
def neighbors_from_neighbors_nodes(G, node):
  list_neighbor = neighbors_nodes(G, node)
  result = []
  for neighbor in list_neighbor:
    list_neighbors_from_neighbors = neighbors_nodes(G, neighbor)
    for n in list_neighbors_from_neighbors:
      # if ((n not in result) and n != node and n not in list_neighbor):
      if(n is not node and n not in list_neighbor):
        result.append(n)
  return result

list_neighbors_from_neighbors_nodes = neighbors_from_neighbors_nodes(graph, "0")
print(list_neighbors_from_neighbors_nodes)
```

```python
## Tm nhng nodes ph bin chung gia 2 node
def popular_between_nodes(G, node1, node2):
  list_neighbor_node1 = neighbors_nodes(G, node1)
  list_neighbor_node2 = neighbors_nodes(G, node2)
  result = []
  for node in list_neighbor_node1:
    if (node in list_neighbor_node2):
      result.append(node)
  return result

popular = popular_between_nodes(graph, "0", "1")
print(popular)
```

```python
## m s ln xut hin ca cc nodes popular vi node cn tm
def count_popular_between_nodes(G, node):
  result = {}
  list_nodes = [n for n in G.nodes()]
  # print(list_nodes)
  # print(node)
  list_node_remove=[]
  for n in list_nodes:
    if(n is not node):
      list_node_remove.append(n)
  # print(list_node_remove)
  neighbors_node=neighbors_nodes(G, node)
  for n in list_node_remove:
    popular_nodes = popular_between_nodes(G,node,n)
    count = len(popular_nodes)
    if( n not in neighbors_node and count>0 ):
      result[n] = len(popular_nodes)
  return result

number_popular_between_nodes=count_popular_between_nodes(graph, "0")
print(number_popular_between_nodes)
```

```python
## Sp xp s ln xut hin ca cc nodes popular vi node cn tm t cao ti thp
def sorted_list(dic):
  so = sorted(dic.items(), key=lambda lam: lam[1],reverse=True)
  result= []
  for s in so:
    result.append(s[0])
  return result
print(sorted_list(number_popular_between_nodes))
```

```python
def recommend_popular(G, node):
  popular = count_popular_between_nodes(G,node)
  result=sorted_list(popular)
  return result
print(recommend_popular(graph, '0'))
```

```python
## Algorithm 2
def affect_map(G, node):
  temp=0
  result={}
  dic = count_popular_between_nodes(G,node)
  for d in dic.keys():
    popular = popular_between_nodes(G,d,node)
    for p in popular:
      numNeighbor = len(neighbors_nodes(G,p))
      temp += (float(1)/numNeighbor)
    result[d]=temp
    temp=0
  return result

print(affect_map(graph,"0"))
```

```python
def recommend_affect(G, node):
  map = affect_map(G, node)
  return sorted_list(map)

print(recommend_affect(graph, "0"))
```

```python
## Tm v tr ca vic gi  kt bn
def find_index(node, list_node):
  result=0
  if (list_node.__contains__(node) == False):
    return result
  else:
    for n in list_node:
      result += 1
      if n==node:
        break
    return result
```

```python
## Tnh v tr ca vic gi  kt bn trung bnh v m s ln th
def average_index(list_node1, list_node2,node1, node2):
  count=float(0)
  average=float(0)
  index1=find_index(node1, list_node2)
  index2=find_index(node2, list_node1)
  if (index2>0.0 and index1>0.0):
    count+=1.0
    average=(index1+index2)/2.0
  return count, average
```

```python
def popular_index(G, node1, node2):
  list_node1=recommend_popular(G, node1)
  list_node2=recommend_popular(G,node2)
  return average_index(list_node1, list_node2,node1, node2)

def affect_index(G, node1, node2):
  list_node1=recommend_affect(G, node1)
  list_node2=recommend_affect(G,node2)
  return average_index(list_node1, list_node2,node1, node2)


```

```python
import random
def comparing(G):
  index_popular = float(0)
  index_effect = float(0)
  count_popular=float(0)
  count_effect= float(0)
  a=[]
  a1=[]
  b=[]
  b1=[]

  for i in range(100):
    node1=random.choice(list(G.nodes.keys()))
    node2=random.choice(list(neighbors_nodes(G,node1)))
    G.remove_edge(node1, node2)
    count_p, index_p = popular_index(G, node1, node2)
    count_popular+=count_p
    index_popular+=index_p
    a.append(index_p)
    a1.append(count_popular)
    count_e, index_e=affect_index(G,node1, node2)
    count_effect+=count_e
    index_effect+=index_e
    b.append(index_e)
    b1.append(count_effect)
    G.add_edge(node1, node2)
  
  result_index_popular = index_popular/count_popular
  result_index_effect=index_effect/count_effect

  print("Average NOCF: ", result_index_popular)
  print("Average IS: ", result_index_effect)

  if (result_index_popular > result_index_effect):
    plt.plot(a1, a, label = "Number of common friends")
    plt.plot(b1, b, label = "Influence score")
    plt.xlabel('S ln test')
    plt.ylabel('Bc xp hng')
    plt.legend()
    plt.show()
    print("IS method is better")
  else:
    plt.plot(a1, a, label = "Number of common friends")
    plt.plot(b1, b, label = "Influence score")
    plt.xlabel('S ln test')
    plt.ylabel('Bc xp hng')
    plt.legend()
    plt.show()
    print("NOCF method is better")
```
