```python

!pip install -q xlrd
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import apriori
OnlineRetail = pd.read_excel('/content/OnlineRetail.xlsx')
OnlineRetail.head()
```

```python
!pip install mlxtend --upgrade
from mlxtend.frequent_patterns import fpgrowth
```

```python
OnlineRetail['Description'] = OnlineRetail['Description'].str.strip()
OnlineRetail
```

```python
OnlineRetail['InvoiceNo'] = OnlineRetail['InvoiceNo'].astype('str')
```

```python
OnlineRetail = OnlineRetail[~OnlineRetail['InvoiceNo'].str.contains('C')]
OnlineRetail
```

```python
OnlineRetail.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
OnlineRetail
```

```python
## Portugal
basket_Portugal = (OnlineRetail[OnlineRetail['Country'] =="Portugal"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket_Portugal.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_Portugal_sets = basket_Portugal.applymap(encode_units)
basket_Portugal_sets.drop('POSTAGE', inplace=True, axis=1)
basket_Portugal_sets
```

```python
## Apriori
apriori(basket_Portugal_sets, min_support=0.05, use_colnames=True)
```

```python
## FPGrowth
fpgrowth(basket_Portugal_sets, min_support=0.05, use_colnames=True)
```

```python
## Association_rules with Portugal
frequent_sets = apriori(basket_Portugal_sets, min_support=0.05, use_colnames=True)
asso_rules = association_rules(frequent_sets, metric="lift", min_threshold=1)
asso_rules.head()
```
